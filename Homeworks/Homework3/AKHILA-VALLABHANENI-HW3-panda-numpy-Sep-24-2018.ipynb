#### Scalar Product / Dot Product
This function returns the dot product of two arrays. For 2-D vectors, it is the equivalent to matrix multiplication. For 1-D arrays, it is the inner product of the vectors. For N-dimensional arrays, it is a sum product over the last axis of a and the second-last axis of b.

```python
dot =np.dot(a,b)
dot
```

```python
a_modulus = np.sqrt((a*a).sum())
b_modulus = np.sqrt((b*b).sum())
cos_angle = dot / a_modulus / a_modulus # returns cosine of angle between a and b
cos_angle
```

```python
cos_angle * 360 / 2 / np.pi # returns angle in degrees
```

### Fancy Indexing
If arrays are indexed by using boolean or integer arrays (masks). The result will be a copy and not a view.

```python
C = np.array([123,188,190,99,77,88,100])
A = np.array([4,7,2,8,6,9,5])
R = C[A<=5]
print(R)
## Indexing with an Integer Array
print(C[[0, 2, 3, 1, 4, 1]])
```

### Non zero
an ndarray a both numpy.nonzero(a) and a.nonzero() return the indices of the elements of a that are non-zero. The indices are returned as a tuple of arrays, one for each dimension of 'a'. The corresponding non-zero values can be obtained 

```python
nz = np.array([[0, 2, 3, 0, 1],
              [1, 0, 0, 7, 0],
              [5, 0, 0, 1, 0]])
print(nz.nonzero())

#If you want to group the indices by element, you can use transpose :transpose(nonzero(nz))
np.transpose(nz.nonzero())
```

The function 'nonzero' can be used to obtain the indices of an array, where a condition is True. In the following script, we create the Boolean array B >= 42

```python
B = np.array([[42,56,89,65],
              [99,88,42,12],
              [55,42,17,18]])
print(B >= 42)
```

### GroupBy: Split, Apply, Combine

```python
#Groupby on type_1 and sort by height in decending order.

c = poke_data.groupby('type_1').sum()
c = c.sort_values(['height'], ascending=False)
c
```

As with the GroupBy object, no computation is done until we call some aggregate on the object:

```python
poke_data.groupby('type_2')['weight']
```

```python
poke_data.groupby('type_2')['weight'].median()
```

### Iteration over groups
The GroupBy object supports direct iteration over the groups, returning each group as a Series or DataFrame

```python
for (method, group) in poke_data.groupby('type_2'):
    print("{0:30s} shape={1}".format(method, group.shape))
```

### Dispatch methods
We can use the describe() method of DataFrames to perform a set of aggregations that describe each group in the data.

```python
poke_data.groupby('type_2')['weight'].describe().unstack()
```

### Aggregation
aggregate() can take a string, a function, or a list thereof, and compute all the aggregates at once.

```python
poke_data.groupby('type_2').aggregate([min, np.median, max])
```

Another useful pattern is to pass a dictionary mapping column names to operations to be applied on that column.

```python
#returns the type of pokemon with minimum height and maximum weight
poke_data.groupby('type_2').aggregate({'height': 'min',
                             'weight': 'max'})
```

### Filtering
A filtering operation allows you to drop data based on the group properties.

```python
#returns only pokemons whose standard deviation of its height is greater than 150.
def func(x):
    return x['height'].std() > 150
display(poke_data.groupby('type_2').filter(func))
```
